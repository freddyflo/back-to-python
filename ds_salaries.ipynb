```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
clean_df = df.dropna()

# Which college major has the lowest starting salary and how much do graduates earn after university?

min_salaray = clean_df['Starting Median Salary'].min()
clean_df['Starting Median Salary'].idxmin()
college_major = clean_df['Undergraduate Major'].loc[49]

print(f"Minimum Salray: {min_salaray}\nCollege Major: {college_major}" )
```

```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
clean_df = df.dropna()

# Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?

min_salaray = clean_df['Mid-Career Median Salary'].min()
position = clean_df['Mid-Career Median Salary'].idxmin()
college_major = clean_df['Undergraduate Major'].loc[position]

print(f"Minimum Salaray: {min_salaray}\nCollege Major: {college_major}" )
```

```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
clean_df = df.dropna()

# What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

max_salary = clean_df['Mid-Career Median Salary'].max()
position = clean_df['Mid-Career Median Salary'].idxmax()
college_major = clean_df['Undergraduate Major'].loc[position]

print(f"Maximum Salaray: {max_salary}\nCollege Major: {college_major}" )
```

```python
import pandas as pd
df = pd.read_csv('salaries_by_college_major.csv')
clean_df = df.dropna()

# A low-risk major is a degree where there is a small difference between the lowest and highest salaries.
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```
